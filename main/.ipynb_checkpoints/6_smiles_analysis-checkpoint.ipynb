```python
import warnings
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Draw
from IPython.display import display
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('../data/LC_ATB_Foc_out_RF_no_salt.csv')
display(df.head())
print(df.shape)
```

```python
def remove_square_brackets(smiles):
    return smiles.replace("[", "").replace("]", "")

def process_smiles(df):
    df['docking_SMILES'] = df['SMILES'].apply(remove_square_brackets)
    return df

df_processed = process_smiles(df)
df.drop(columns='SMILES', inplace=True)
df.rename(columns={'docking_SMILES':'SMILES'}, inplace=True)
# df.to_csv('176_smiles.csv', index=False)
display(df_processed.head())
print(df_processed.shape)
print(df_processed.columns)
```

```python
df_out = pd.read_csv('output.csv')

with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=FutureWarning)
    df_out = pd.read_csv('output.csv')
    df_out['new_name'] = df_out['Name'].str.replace(r'output_\d+_','')
    df_out['new_name'] = df_out['new_name'].str.rsplit('_', 1).str[0]
    df_out[['ligand_1']] = df_out['new_name'].str.rsplit('_', 2, expand=True)[[0]]
    df_out['ligand_2'] = df_out['new_name'].str.rsplit('_', 2).str[1:].str.join('_')
    df_out.drop(columns='new_name', inplace=True)

display(df_out.head())
print(df_out.columns)
```

```python
ligand_1 = pd.read_csv('brd4_480k.csv')
ligand_2 = pd.read_csv('e3_vhl_5k.csv')

for index, row in df_out.iterrows():
    match_1 = ligand_1[ligand_1['Name'] == row['ligand_1']]
    if not match_1.empty:
        df_out.loc[index, 'ligand_1_SMILES'] = match_1['SMILES'].values[0]
        
    match_2 = ligand_2[ligand_2['Name'] == row['ligand_2']]
    if not match_2.empty:
        df_out.loc[index, 'ligand_2_SMILES'] = match_2['SMILES'].values[0]

df_out = df_out.reindex(columns=['Name', 'SMILES', 'Affinity', 'Efficiency', 'ligand_1', 'ligand_1_SMILES', 'ligand_2', 'ligand_2_SMILES'])

display(df_out.head())
print(df_out.shape)
```

```python
def extract_linker(row):
    smiles1 = row['ligand_1_SMILES']
    smiles2 = row['ligand_2_SMILES']
    smiles_comb = row['SMILES']
    
    mol1 = Chem.MolFromSmiles(smiles1)
    mol2 = Chem.MolFromSmiles(smiles2)
    mol_comb = Chem.MolFromSmiles(smiles_comb)
    
    if mol_comb.HasSubstructMatch(mol1) and mol_comb.HasSubstructMatch(mol2):
        indices_to_remove = set(mol_comb.GetSubstructMatches(mol1)[0]) | set(mol_comb.GetSubstructMatches(mol2)[0])
        
        remaining_part = Chem.RWMol(mol_comb)
        for idx in sorted(indices_to_remove, reverse=True):
            remaining_part.RemoveAtom(idx)
        
        remaining_part = remaining_part.GetMol()
        smiles_remaining = Chem.MolToSmiles(remaining_part)
        return smiles_remaining
    else:
        return None

df_out['linker'] = df_out.apply(extract_linker, axis=1)

df_out.to_csv('difflinker_output.csv', index=False)
display(df_out.head())
print(df_out.shape)
```

```python
def draw_molecule_from_df(df, smiles_column, index, img_size=(800, 400)):
    smiles = df.loc[index, smiles_column]
    mol = Chem.MolFromSmiles(smiles)
    
    if mol is not None:
        img = Draw.MolToImage(mol, size=img_size)
        display(img)
    else:
        print(f"Invalid SMILES at index {index}.")


def get_smiles_from_df(df, smiles_column, index):
    smiles = df.loc[index, smiles_column]
    return smiles

smiles_column_name = 'SMILES'
row_index = 432

draw_molecule_from_df(df, smiles_column_name, row_index, img_size=(800, 400))
smiles = get_smiles_from_df(df, smiles_column_name, row_index)
print("SMILES:", smiles)
```

```python
from rdkit import Chem
from rdkit.Chem import Draw
import h5py
import numpy as np

def smiles_to_molgraph(smiles):
    mol = Chem.MolFromSmiles(smiles)
    return mol

def save_molgraph_to_h5(mol, h5_filename):
    atom_features = np.array([atom.GetAtomicNum() for atom in mol.GetAtoms()])
    bond_features = np.array([bond.GetBondTypeAsDouble() for bond in mol.GetBonds()])
    
    with h5py.File(h5_filename, 'w') as f:
        f.create_dataset('atom_features', data=atom_features)
        f.create_dataset('bond_features', data=bond_features)

smiles = 'CCO'
mol_graph = smiles_to_molgraph(smiles)
save_molgraph_to_h5(mol_graph, 'mol_graph.h5')
```

```python
def load_molgraph_from_h5(h5_filename):
    with h5py.File(h5_filename, 'r') as f:
        atom_features = f['atom_features'][:]
        bond_features = f['bond_features'][:]
    
    return atom_features, bond_features

atom_features, bond_features = load_molgraph_from_h5('mol_graph.h5')
print("Atom features:", atom_features)
print("Bond features:", bond_features)
```
